```python
df['Age'] = df['Age'].interpolate()
df['Age'].isnull().sum()
```

```python
df['Cabin'] = df['Cabin'].fillna(method="ffill")
df['Cabin'].isnull().sum()
```

```python
df['Cabin'] = df['Cabin'].fillna(method="bfill")
df['Cabin'].isnull().sum()
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
df['Embarked'].isnull().sum()
```
